```python
import pandas as pd
```

```python
data = pd.read_excel("D:\\OneDrive\\02_Programs\\2021-Data-competition-Shangdong\\烟台1-创意赛-城市新能源车充电桩安装位置规划\\表1烟台地区现有充电桩信息.xlsx")
data_2 = pd.read_excel("D:\\OneDrive\\02_Programs\\2021-Data-competition-Shangdong\\烟台1-创意赛-城市新能源车充电桩安装位置规划\\表2烟台地区充电站充电次数及充电量信息.xlsx")
```

```python
data.head()
```

```python
data_2.drop(columns=['序号'], inplace=True)
data_2.head()
```

```python
xx = data.groupby('充电站名称', sort=False).sum()
xx.head()
```

```python
data_processing = pd.DataFrame(data['充电站名称'].value_counts(sort=False))
data_processing.columns = ['Number of charging piles']
data_processing.head()
```

```python
xx['序号'] = data_processing['Number of charging piles']
xx.reset_index(inplace=True)
xx.head()
```

```python
xx['纬度'] = xx['纬度']/xx['序号']
xx['经度'] = xx['经度']/xx['序号']
xx.head()
```

```python
xx.rename(columns={'序号':'Number of charging piles','充电站名称':'充电站' }, inplace=True)
xx.to_csv('data_process.csv', encoding='utf_8_sig')
```

```python
xx.head() 
```

```python
xx_2 = pd.merge(data_2, xx)
xx_2.head()
```

```python
xx_2.to_csv('data_emerge_all.csv', encoding='utf_8_sig')
```
